```python
import re
df = tt.copy()

df['year'] = ""

def process_title(row):

    match = re.search(r'\((.*?)\)', row['title'])
    if match:
        bracket_content = match.group(1)
        
        year_match = re.search(r'(\d{4})(?:–(\d{4})?|\s*–\s*)?', bracket_content)
        if year_match:
            start_year = year_match.group(1)
            end_year = year_match.group(2) if year_match.group(2) else ''
            
            if end_year:
                row['year'] = f"{start_year}–{end_year}"
            else:
                row['year'] = start_year

    row['title'] = re.sub(r'\s*\(.*?\)', '', row['title']).strip()
    return row

df = df.apply(process_title, axis=1)
```

```python
from datetime import datetime

def clean_release_date(date_str):
    if pd.isna(date_str) or date_str == "":
        return ""
    
    date_str = re.sub(r'\s*\(.*?\)', '', date_str).strip()
    
    try:
        date_obj = datetime.strptime(date_str, "%B %d, %Y")
        return date_obj.strftime("%Y-%m-%d")
    except ValueError:
        return date_str

df['release_date'] = df['release_date'].apply(clean_release_date)
```

```python
ad = pd.read_csv("ain.csv", dtype={'imdb_id': str})
pr = pd.read_csv("tv_series.csv", dtype={'imdb_id': str})

merged = pd.merge(pr, ad[['imdb_id', 'additional_info']], on='imdb_id', how='left')

merged['production_details'] = merged['production_details'].fillna('').astype(str).str.strip()
merged['additional_info'] = merged['additional_info'].fillna('').astype(str).str.strip()

def combine_text(prod, add):
    if prod and add:
        return f"{prod}; {add}"
    return prod or add  

merged['production_details'] = merged.apply(
    lambda row: combine_text(row['production_details'], row['additional_info']),
    axis=1
)

merged.drop(columns=['additional_info'], inplace=True)
merged.to_csv("merged_output.csv", index=False)
```
